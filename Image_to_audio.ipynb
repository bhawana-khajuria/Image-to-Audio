```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('Image.webp')
print(img)
```

Your image should carry some text

```python
result = pytesseract.image_to_string(img)
print(result)

if result:  # Check if result is not empty
    tts = gtts.gTTS(result)
    tts.save("hello.mp3")
    print("Audio file 'hello.mp3' created successfully.")
else:
    print("No text detected in the image to convert to speech.")
```
